```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

options = Options()
options.add_argument('--headless') 
options.add_argument('--disable-gpu') 

# Inicializar o driver do Selenium
driver = webdriver.Chrome(options=options)
url = 'https://pidigital.pi.gov.br/app/catalog/list/'

# Navegar até a página
driver.get(url)

# Lista para armazenar os resultados
resultados = []

# Esperar até que os cards de serviço estejam carregados
wait = WebDriverWait(driver, 10)
cards = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "xvia-listing-card")))

# Iterar sobre cada card
for card in cards:
    # Extrair o título do serviço
    servico = card.find_element(By.CLASS_NAME, "xvia-listing-card__title").text
    
    # Extrair a descrição do serviço
    descricao = card.find_element(By.CLASS_NAME, "xvia-listing-card__subtitle").text
    
    # Extrair o URL do serviço (link do botão)
    botao = card.find_element(By.CSS_SELECTOR, "button.xvia-listing-card__bottom-button")
    url_servico = botao.get_attribute("onclick")  # Pode ser necessário ajustar dependendo de como o link está implementado
    
    # Adicionar os dados extraídos à lista de resultados
    resultados.append({
        "servico": servico,
        "descricao": descricao,
        "url_servico": url_servico
    })

# Imprimir os resultados
for resultado in resultados:
    print(f"Serviço: {resultado['servico']}")
    print(f"Descrição: {resultado['descricao']}")
    print(f"URL do serviço: {resultado['url_servico']}")
    print("---")

# Fechar o navegador
driver.quit()
```
