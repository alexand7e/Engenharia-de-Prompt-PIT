````python
# def convert_dict(text):
#     try:
#         text = text.replace('null', '""').replace('```', '')
#         return ast.literal_eval(text)
#     except (SyntaxError, ValueError):
#         return None

def convert_dict(input_string):
    start = input_string.find('{')
    end = input_string.rfind('}') + 1
    
    if start != -1 and end != -1:
        dict_string = input_string[start:end]
        try:
            result_dict = eval(dict_string)
            if isinstance(result_dict, dict):
                return result_dict
        except:
            pass
    return {}
````

````python
def analyze_problem(analyzer: GeminiAnalyzer, problem_text: str, escolaridade: str, experiencia: str):
    instruction = f'''
    <perfil aluno>
    Escolaridade: {escolaridade}
    Experiência com IA: {experiencia}
    </perfil aluno>
    <projeto>
    {problem_text}
    </projeto>
    <instrução>
    User: Analise o conteúdo dos projetos e faça uma análise, criando os seguintes campos resultado da sua análise:
    - dificuldade
    - resumo em tópicos
    - viabilidade
    - estratégia e passo a passo de implementação
    A saída deve ser um dicionário Python
    </instrução>
    <resultado>
    IA: ```
    {{
    "dificuldade": "média",
    "resumo": ["Analisar dados", "Desenvolver modelo", "Testar resultados", "Funcionalidade X"],
    "viabilidade": "média",
    "estratégia": "Você deve utilizar a ferramente X (www.exemplo.com), mapear e documentar seus livros. Automaticar a transcrição e treinar o modelo Y local."
    }}
    ```
    </resultado>
    '''

    print(f"Prompt: {instruction}")
    resposta, _ = analyzer.generate_google_response(instruction)
    
    return resposta

def main(data):
    analyzer = GeminiAnalyzer()
    results = []
    for index, row in data.iterrows():
        result = analyze_problem(analyzer, row['Projeto'], row['Escolaridade'], row['Experiência com IA'])
        print(result)
        results.append(result)
        new_fields = convert_dict(result)
        print(new_fields)

        for key, value in new_fields.items():
            data.loc[index, key] = str(value)

        sleep(3)
    return results, data
````
